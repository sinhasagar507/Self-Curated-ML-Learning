```python
# Reframe precision of regression forecast

## Load dataset 
data = pd.read_csv("../input/daily-minimum-temperatures-in-melbourne/daily-minimum-temperatures-in-me.csv", header=0, index_col=0, parse_dates=True, squeeze=True, on_bad_lines='warn')
values = data.values
df = pd.DataFrame(values)
df.columns = ['t+1']
df['t+1'] = df['t+1'].apply(lambda x: x.replace('?', ''))
df['t+1'] = df['t+1'].astype(float)

## Create a lagged dataset 
df = pd.concat([df.shift(1), df], axis=1)
df.columns = ['t', 't+1']
df = df.dropna()

## Create a rolling mean for the past fortnight 
df['rolling'] = df['t+1'].rolling(window=14).mean()
df = df.dropna()
```
